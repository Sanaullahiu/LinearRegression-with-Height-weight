# Import  Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import csv File

```python
df=pd.read_csv('Height-Weight Data.csv')
df.head()
```

# Define Data

```python
x=df[['Weight']]
x.head()
```

```python
y=df[['Height']]
y.head()
```

# Create Scatter Plot with labeling

```python
plt.figure(figsize=(10,8))
plt.scatter(x, y,color="green", marker="*")

font1={'family':'arial','size':'14','color':'black'} #not mandatory,but it providing nice view of tittle and label
font2={'family':'arial','size':'18','color':'black'}

plt.xlabel("Weight",fontdict=font1)
plt.ylabel("Height",fontdict=font1)
plt.title("Comparison of Weight & Height",fontdict=font2)
```

# Seperate training and testing data

```python
from sklearn.model_selection import train_test_split 
xtrain,xtest,ytrain,ytest= train_test_split (x,y, test_size=0.3)
```

```python
xtrain.head(),"  ", ytrain.head(), "  ", xtest.head(), " ", ytest.head()
```

# Model-Selection

```python
from sklearn.linear_model import LinearRegression 
reg=LinearRegression()
reg.fit(xtrain,ytrain)
```

# find out Coefficient and intercept value

```python
reg.coef_, " ", reg.intercept_
```

# Determine any pridict value 

```python
reg.predict([[1200]]) # we can also predict any value with input in predict funtion
```

# Create best fit line

```python
plt.plot(df.Weight,reg.predict(df[['Weight']]))
```

# Fit the strateline with scatter value

```python
plt.figure(figsize=(10,8))
plt.scatter(xtrain, ytrain,color="green", marker="*")

font1={'family':'arial','size':'14','color':'black'} #not mandatory,but it providing nice view of tittle and label
font2={'family':'arial','size':'18','color':'black'}

plt.xlabel("Weight",fontdict=font1)
plt.ylabel("Height",fontdict=font1)
plt.title("Comparison of Weight & Height",fontdict=font2)

plt.plot(df.Weight,reg.predict(df[['Weight']]))
```
